# 04. Final analysis

Aggregate all detection / mitigation runs into the figures and tables that go into the thesis.

```python
import json, glob
import pandas as pd
rows = []
for path in sorted(glob.glob('../results/detection/*_metrics.json')):
    with open(path) as f:
        for method, m in json.load(f).items():
            rows.append({'file': path.split('/')[-1], 'method': method, **m})
df = pd.DataFrame(rows)
df.head()
```
