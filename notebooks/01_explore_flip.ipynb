# 01. Explore FLIP outputs

Inspect the precomputed FLIP soft labels, look at score distributions, and confirm the (source -> target) classes that dominate.

```python
import sys, pathlib
ROOT = pathlib.Path('..').resolve()
if str(ROOT) not in sys.path: sys.path.insert(0, str(ROOT))
import numpy as np, torch, torchvision
from src.attacks.flip_wrapper import score_flip_candidates
```

```python
soft_labels_path = '../external/FLIP/precomputed/cifar10_sinusoidal_soft_labels.pt'
soft = torch.load(soft_labels_path, weights_only=False).numpy() if hasattr(torch.load(soft_labels_path, weights_only=False), 'numpy') else np.asarray(torch.load(soft_labels_path, weights_only=False))
cifar = torchvision.datasets.CIFAR10('../data/raw', train=True, download=True)
labels = np.asarray(cifar.targets, dtype=np.int64)
scores, targets = score_flip_candidates(soft, labels)
print('Top scores:', np.sort(scores)[-10:])
print('Most common (src,tgt):')
from collections import Counter
top = np.argsort(scores)[-1000:]
print(Counter(zip(labels[top].tolist(), targets[top].tolist())).most_common(5))
```
